## Visualize the Data

```python
# this visualization helps me see what data points are outliers and i can run a filter to view them
sns.set()
features_plot = sns.pairplot(df, hue='brand', size=2.5);
features_plot.savefig('../figures/feature_pairplot.png')
plt.show()
```

```python
# histograms

df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
sns.regplot(df['mileage(mi)'], df['price(£)'])
```

```python
df['brand'].value_counts().plot(kind='bar', figsize=(10,4))
```

```python
# filters to inspect the outliers

# highest mileage car
filt = df['mileage(mi)']== df['mileage(mi)'].max()

# cheapest car
filt2 = df['price(£)'] == df['price(£)'].min()

# engine size
filt3 = df['engine_size(cc)'] == df['engine_size(cc)'].max()

df[filt3]
```

```python
# most expensive 2020 car brand
df.query('year==2020').groupby("brand").agg(avg_price=("price(£)","mean")).sort_values("avg_price", ascending=False).round(2)
```

```python
# most expensive cars brand in my data set
df.groupby("brand").agg(avg_price=("price(£)","mean")).sort_values("avg_price", ascending=False).round(2).head()
```

```python
# amount of brands
len(list(df['model'].unique()))
```

```python
# Median price per brand (use this as an app to recommend brands based of users budget range)
brand_median_price = df.groupby('brand').agg(Price= ("price(£)", "median"),
                                             Quantity= ("model", "count"))
brand_median_price
```

```python
# gives average price of each brand and their model type
median_brand_price = df.groupby(['brand', 'model']).agg(
                        Price = ("price(£)", "median"),
                        Quantity = ("brand", "count"),
                        Model_Avg_Miles = ("mileage(mi)", "median"))

median_brand_price
```
