## Dropping outliers within 6 standard deviations:

In this section I will manage my outliers by dropping outliers from the price, engine_size, mileage, and year columns beyond 6 sigma's of the means of these columns.

```python
df_drop_outliers = df[(np.abs(stats.zscore(df[['price(£)', 'engine_size(cc)', 'mileage(mi)','year']])) < 6).all(axis=1)]
```

```python
sns.set()
features_plot = sns.pairplot(df_drop_outliers, hue='brand', size=2.5);
features_plot.savefig('../figures/feature_pairplot.png')
plt.show()
```

That seems to have done the trick, my extreme outliers are gone I still have a few outliers that I want my model to try to catch such as the expensive classical sports cars under the year 2000 etc.

## Feature Engineering

In this section I will be creating new features for my data set based off my car knowledge. I will make features such as:
1. Is the cars brand luxury or not?
2. Is the car brand new? (low mileage)
3. is the car old? (high mileage)
4. is the car considered a sports car? (big engine, 2 doors, coupe body style)

### Luxury brand
Based off popular culture, good marketing, manufacturing quality, and control of inventory, certain brands of the years have gained a luxurious image. These brands can be reflected within my data set as expensive brands from my pivot tables above but since my data set does not reflect the entire population of cars and brands, I will use my personal knowledge on cars to define this category.

```python
# Defined by common knowledge of car brands
luxury_brands = ['Ferrari', 'Mclaren', 'Aston-Martin', 'Bentley', 'Porsche', 'Maserati', 'Land-Rover', 'Mercedes-Benz', 'Alfa-Romeo', 'Jaguar', 'Audi', 'BMW']
```

I will use this list to compare with my data sets brand column and add a 1 for anytime one of the cars in my data frame show up on this list and a 0 when they don't.

```python
# container to hold binary 1 or 0
luxury = []

for item in df_drop_outliers['brand']:
    if item in luxury_brands:
        luxury.append(1)
    else: 
        luxury.append(0)
        
# Create new column for my binary values        
df_drop_outliers['luxury_brand'] = luxury

# show the addition
df_drop_outliers.head()
```

```python
# check feature value counts
print("There are :", df_drop_outliers['luxury_brand'].value_counts()[1], "luxury cars")
print("out of:", len(df_drop_outliers), "cars")
```

### Brand New
I consider a car with less than 10k miles on the odometer as brand new. These cars are from the UK and on average british people drive 7.9k miles per year according to my [source](https://www.nimblefins.co.uk/average-annual-mileage-cars-england-down-%E2%80%93-are-we-really-driving-less).


```python
# container to hold the binary 1 or 0
brand_new = []

for item in df_drop_outliers['mileage(mi)']:
    if item <= 10000:
        brand_new.append(1)
    elif item > 10000:
        brand_new.append(0)
        
# create the column        
df_drop_outliers['brand_new'] = brand_new
```

```python
# show the addition
df_drop_outliers.head()
```

```python
# check feature value counts
print("There are :", df_drop_outliers['brand_new'].value_counts()[1], "brand new cars")
print("out of:", len(df_drop_outliers), "cars")
```

### Old
To create this feature, I have to establish a few parameters of what I consider new to be. Just like my brand new feature I know that the data set is of cars within England, I know that on average, british people tend to drive 7.9k miles per year according to my [source](https://www.nimblefins.co.uk/average-annual-mileage-cars-england-down-%E2%80%93-are-we-really-driving-less). <br>
I have decided to set my defining mileage limit for this feature at cars with over 100k miles on the odometer.

```python
# list to hold the binaries
old = []

for item in df_drop_outliers['mileage(mi)']:
    if item > 100000:
        old.append(1)
    elif item <= 100000:
        old.append(0)
        
# create the feature
df_drop_outliers['old']= old
```

```python
# show the addition
df_drop_outliers.head()
```

```python
# check the car counts
print("There are :", df_drop_outliers['old'].value_counts()[1], "cars considered old")
print("out of:", len(df_drop_outliers), "cars")
```

### Sports Car
We know just by media, sports cars are often saught after. I will define my cars based on a few characteristics that sports cars tend to have. Big engines and 2 doors!

```python
# container to house binary
sports_car = []

for item in df_drop_outliers.values:
    if item[4] > 2500 and item[3] == "Coupe":
        sports_car.append(1)
    else:
        sports_car.append(0)

# Create the column
df_drop_outliers['sports_car']= sports_car
```

```python
# View the addition
df_drop_outliers.head()
```

```python
# check the car counts
print("There are :", df_drop_outliers['sports_car'].value_counts()[1], "sports cars")
print("out of:", len(df_drop_outliers), "cars")
```

## Correlation of new features
What does the correlation between my new features and price look like?

```python
new_features = df_drop_outliers.loc[:,['price(£)', 'luxury_brand','brand_new','sports_car','old']]
heat_map(new_features, (10,7))
```
